```python
import pandas as pd
marketing_cl_pd=pd.read_csv('marketing_campaign_cleaned.csv')
marketing_cl_pd.head()
```

```python
marketing_cl_pd['Education'].value_counts()
```

```python
marketing_cl_pd['Marital_Status'].value_counts()
```

```python
marketing_cl_pd_encoded=pd.get_dummies(marketing_cl_pd,columns=['Education','Marital_Status'],drop_first=True)
marketing_cl_pd_encoded.head()
marketing_cl_pd_encoded.dtypes
```

```python
marketing_cl_pd_encoded=marketing_cl_pd_encoded.astype(int)
marketing_cl_pd_encoded.dtypes
```

```python
marketing_cl_pd_encoded.to_csv('marketing_campaign_cleaned_encoded.csv',index=False)
```
